<p><span style="font-size: 14pt; font-family: verdana, geneva;"><strong>Word Embeddings Vs Pretrained Models</strong></span></p>
<p><span style="font-size: 11pt; font-family: verdana, geneva;">When training NLP models, traditional methods include using pretrained word embeddings as the first layer in the neural network. Word Embeddings such as word2vec and GloVe were popular word embeddings that represented words as a single layer of weights. Once your corpus was transformed into the the word embeddings you had to create your own model and begin training from scratch. This method still requires a large amount of new data and computational time to learn the downstream task.</span></p>
<p><span style="font-size: 11pt; font-family: verdana, geneva;">Using the entire pretrained model as a whole, freezing the weights at each layer, then adding my own network ontop of the pretrained model is the goal of this notebook. By using the pretrained models archetcheture and weights, I get to take advantage of all the training time that the previous research did on large corpus data sets and not having to model the language from scratch. Another advantage in using the model in its entirety is that high level representations present as well as low level representations, allowing for more language understanding throughout.&nbsp;</span></p>
<p>&nbsp;</p>

<p><span style="font-size: 14pt; font-family: verdana, geneva;"><strong>Three Steps to Using a Pretrained Model for NLP Tasks</strong></span></p>
<ol>
<li><span style="font-size: 12pt;">Languague Modeling. Unsupervised Learning is used to train the model on large corpus to learn high and low level features.</span></li>
<li><span style="font-size: 12pt;">Fine-tuning the pre-trained language model on target task data. Model is then used in a supervised manner and trained on a specific NLP task such as sentiment, document classification, etc.</span></li>
<li><span style="font-size: 12pt;">Fine-tuning the classifier. Classifier is fine-tuned on the target task using gradual unfreezing and STLR to preserve low-level representations and adapt high-level ones (shaded: unfreezing stages; black: frozen).</span></li>
</ol>

```python
train_df = pd.read_csv('./Data/train_tab_delimit.txt', delimiter='\t', header=0, encoding = "latin-1")
test_df = pd.read_csv('./Data/test_tab_delimit.txt', delimiter='\t', header=0, encoding = "latin-1")
```

```python
### Simple clean function for tweets
### I did not apply lower case to the tweets due to objective of classifying for or against in stance

def clean_ascii(text):
    # function to remove non-ASCII chars from data
    return ''.join(i for i in text if ord(i) < 128)

train_df['Tweet'] = train_df['Tweet'].apply(clean_ascii)
```

```python
### ULMFit only needs Stance and Tweet with Stance being placed in front of tweet
train = pd.concat([train_df['Stance'], train_df['Tweet']], axis=1)
train.to_csv('./Data/train.csv', index=False, header=False)
```
